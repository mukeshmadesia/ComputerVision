## 1. Import Necessary libraries

```python
from keras.preprocessing.text import Tokenizer
from keras.models import Sequential
from keras.layers import Embedding,LSTM,Dense,Dropout
import requests
```

## 2. Import Data

```python
my_data = """Artificial Intelligence Program is really going good.\n
           My recruiters are giving their best efforts and time in reviewing my profile at the given time.\n
           I completely trust my recruiters but this task is to show my knowledge about the Topic.\n
           Even if I am lagging somewhere I will put efforts to improve that area."""
my_data
```

## 3. Data Preparation

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([my_data])
encoded_data = tokenizer.texts_to_sequences(texts = [my_data])[0]
encoded_data
```

```python
min(encoded_data),max(encoded_data)
```

```python
vocab = len(set(encoded_data)) + 1
vocab
```

## 3.1 Create sequence of words to fit the model with one word as input and one word as output

```python
len(encoded_data)
```

```python
sequences = []

for i in range(1,len(encoded_data)):
    sequence = encoded_data[i-1:i+1]
    sequences.append(sequence)
print('Total number of Sequences : ',len(sequences))
```

```python
sequences
```

### 3.2 Split the Sequences into Training and Test

```python
import numpy as np
sequences = np.array(sequences)
X = sequences[:,0]
y = sequences[:,1]
```

```python
X[:5]
```

```python
y[:5]
```

## 4. Model Building

```python
model = Sequential()
model.add(Embedding(input_dim = vocab, output_dim = 40,input_length = 1))
model.add(LSTM(units = 50,dropout = 0.15))
model.add(Dense(units = 50,activation = 'tanh'))
model.add(Dropout(0.40))
model.add(Dense(units = vocab,activation = 'softmax'))
model.summary()
```

```python
model.compile( optimizer='adam',loss='sparse_categorical_crossentropy',metrics='accuracy')
```

## 5. Model Training

```python
model.fit(x=X,y=y,epochs=1000)
```

## 6. Model Prediction

### Generate new words based on the previous

```python
tokenizer.word_index.items()
```

```python
def generate_text_sequence(model,tokenizer,enter_text,n_pred):
    in_text,result = enter_text,enter_text
    
    for _ in range(n_pred):
        encoded_data = tokenizer.texts_to_sequences(texts = [in_text])[0]
        encoded_data = np.array(encoded_data)
    
        y_pred = model.predict(encoded_data)
        
        #Mapping the predicted word index to the word
        predicted_word = ''
        for word,index in tokenizer.word_index.items():
            if index ==y_pred.all():
                predicted_word = word
                break
        
        in_text,result =  predicted_word, result + ' ' + predicted_word
    return  result       
```

```python
generate_text_sequence(model,tokenizer,'recruiters',3)
```

```python
generate_text_sequence(model,tokenizer,'lagging',2)
```
